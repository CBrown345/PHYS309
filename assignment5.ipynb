1. The hamiltonian operator is an operator related to the total energy of a system. Since the total energy is the kinetic energy plus the potential, the hamiltonian can be written as $\hat{H} = \hat{T} + \hat{V}$ where $\hat{T}$ and $\hat{V}$ represent the kinetic and potential energies respectively. Because kinetic energy is defined as $p^2 /2m$, $\hat{T}$ gets replaced by $\hat{p}^2/2m$ So, the Hamiltonian becomes,
\begin{equation*}
    \hat{H} = -\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2} + \hat{V}
\end{equation*}

3. The eigenstates correspond to the solutions of the harmonic oscillator potential Shroedinger equation. They look like (Griffiths Quantum Mechanics, equation 2.86),
\begin{equation}
    \psi_n (x) = \left( \frac{m\omega}{\pi \hbar}\right)^{1/4}\frac{1}{\sqrt{2^nn!}}H_n(\sqrt{\frac{m\omega}{\hbar}}x)e^{-\frac{m\omega x^2}{2\hbar}}.
\end{equation}
where $H_n$ represents the Hermite polynomials ($n \in \mathbb{Z}, n \geq 1$). Notably, the energy of the $n{\rm th}$ state is $E_n = (n + \frac{1}{2})\hbar \omega$

```python
import numpy as np
import matplotlib.pyplot as plt

from qmsolve import Hamiltonian, SingleParticle, init_visualization, Ã…, eV
```

```python
# Harmonic oscillator potential
def harmonic_oscillator(particle):
    k = 100 * eV / Ã…**2
    return 0.5 * k * particle.x**2

# Define the Hamiltonian
H = Hamiltonian(particles = SingleParticle(), potential = harmonic_oscillator, 
                spatial_ndim = 1, N = 512, extent = 20*Ã…)

#Diagonalize the Hamiltonian and compute the eigenstates
eigenstates = H.solve(max_states = 30)

# Print out the energies of the eigenstates
print(eigenstates.energies) # the printed energies are expressed in eV

# Visualize the Eigenstates
visualization = init_visualization(eigenstates)
visualization.plot_eigenstate(0)
visualization.plot_eigenstate(1)
visualization.plot_eigenstate(2)
visualization.plot_eigenstate(3)
visualization.plot_eigenstate(5)
visualization.plot_eigenstate(7)
```
